# 1

# write a program to generate PDF of uniform distribution function
$$f(x)=\begin{cases} \frac{1}{b-a} & \text{ ; } a\leq x\leq b \\ 0& \text{ ; } otherwise \end{cases} $$

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import uniform
import seaborn as sns
```

```python
a = 10
b = 30
n = 10000
fx_uniform = uniform.rvs(loc=a,scale=b-a,size=n)

ax = sns.distplot(fx_uniform)
ax.set(xlabel='x-->',ylabel='Frequency',title='Uniform Distribution')
```

# 2

# Write a program to generate PDF of Normal Distribution function
$$f(x) = \frac{1}{2 \pi \sigma^{2}}.e^{\frac{-(x-\mu)^{2}}{2\sigma^{2}}}$$

```python
from scipy.stats import norm
mean = 0
std = 1
n = 10000
fx_normal = norm.rvs(size = n,loc = mean,scale = std)

bx = sns.distplot(fx_normal)
bx.set(xlabel = 'x-->',ylabel = 'Frequency-->',title = 'Normal Distribution')
```

# 3

# Write a program to generate PDF of Gamma distribution function.
$$f(x) = \frac{\beta^{\alpha}\; x^{\alpha-1}\;e^{-\beta x} }{\Gamma (a)}$$

```python
from scipy.stats import gamma
mean = 0
std = 1
n = 10000
fx_gm = gamma.rvs(size = n,a = 5,loc = mean,scale = std)

bx = sns.distplot(fx_gm)
bx.set(xlabel = 'x-->',ylabel = 'Frequency-->',title = 'Gamma Distribution')
```

# 4

# Write a program to generate PDF of Exponential distribution function.
$$f(x) = \begin{cases} \lambda e^{-\lambda x} & \text{ if } x\geq 0 \\ 0 & \text{ if } x<0 \end{cases}$$

```python
from scipy.stats import expon
mean = 0
std = 1
n = 10000
fx_ex = expon.rvs(size = n,loc = mean,scale = std)

bx = sns.distplot(fx_ex)
bx.set(xlabel = 'x-->',ylabel = 'Frequency-->',title = 'Exponential Distribution')
```

# 5

# Write a program to generate PDF of Binomial distribution function.
$$f(x) = _{x}^{n}\textrm{C}\; p^{x} (1-p)^{n-x}$$

```python
from scipy.stats import binom
fx_bi = binom.rvs(size = 10000,n = 10,p = 0.8)

bx = sns.distplot(fx_bi)
bx.set(xlabel = 'x-->',ylabel = 'Frequency-->',title = 'Binomial Distribution')
```
