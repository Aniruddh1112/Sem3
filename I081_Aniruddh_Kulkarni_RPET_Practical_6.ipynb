## Q.1

## write a program to calculate fourier transform

```python
xn = eval(input("Enter the input sequence: "))
xn = list(xn)
```

```python
xn
```

```python
Xk = np.fft.fft(xn)
Xk
```

```python
PSD = np.abs(Xk)
PSD
```

```python
plt.stem(PSD)
plt.xlabel('k --->')
plt.ylabel('X(k) --->')
plt.title("Power Spectral Density")
```

## Q.3


## write a program to find the psd of a sine signal

```python
F1 = 20
F2 = 40
fs = 500
t = np.arange(0, 2, (1/fs))
```

```python
plt.figure(figsize = (15, 3))
x1 = np.sin(2*np.pi*F1*t)
x2 = np.sin(2*np.pi*F2*t)
plt.plot(t, x1)
```

```python
plt.figure(figsize = (15,3))
xn = x1 + 2*x2
plt.plot(t, xn)
```

```python
xn.shape
```

```python
from IPython.display import Audio
Audio(data = xn, rate = fs)  
```

```python
Xk = np.fft.fft(xn)
PSD = np.abs(Xk)
```

```python
N = len(xn)
```

```python
K = np.arange(N)*(fs/N)
```

```python
plt.figure(figsize = (15, 3))
plt.stem(K[0:int(N/2)],PSD[0:int(N/2)])
plt.xlabel("Frequency ---->")
plt.ylabel("Magnitude ---->")
plt.title("Power Spectral Density")
```

## Q.4

## write a program to calculate psd of a noise/random signal

```python
n = np.random.normal(-1,1,len(xn))
```

```python
n.shape
```

```python
yn = xn + n
```

```python
plt.figure(figsize = (15,3))
plt.plot(t,yn)
```

```python
ryn = np.correlate(yn, yn, mode = 'same')
```

```python
plt.figure(figsize = (15,3))
plt.plot(ryn)
```

```python
Xk = np.fft.fft(ryn)
PSD = np.abs(Xk)
```

```python
K = np.arange(N)*(fs/N)
```

```python
plt.figure(figsize = (15, 3))
plt.stem(K[0:int(N/2)],PSD[0:int(N/2)])
plt.xlabel("Frequency ---->")
plt.ylabel("Magnitude ---->")
plt.title("Power Spectral Density")
```
